```python
# prompt: df.columns.str.replace(for x in df 'Unnamed ', 'df[x]row[1] or row[2])

import pandas as pd
# Assuming df is already defined and loaded as in the previous code

# Method 1: Using a loop and direct assignment (less efficient for large DataFrames)
for col in df.columns:
    if 'Unnamed' in col:
        new_col_name = ''
        # Accessing the first two rows (index 0 and 1)
        # Note: This assumes your first row is at index 0 and has text you want
        if pd.notna(df[col].iloc[0]) and len(str(df[col].iloc[0])) > 0:
          new_col_name = str(df[col].iloc[0])
        elif pd.notna(df[col].iloc[1]) and len(str(df[col].iloc[1])) > 0:
            new_col_name = str(df[col].iloc[1])
        else:
          new_col_name = col  # Keep original name if both rows are NaN or empty
        df.rename(columns={col: new_col_name}, inplace=True)


# Method 2 (more efficient) Using pandas .replace() method
# df.columns = df.columns.str.replace('Unnamed', ' ') # replace 'Unnamed' by nothing
# print(df.columns)
df.columns
```
